```python
import sqlite3
conn=sqlite3.connect('my_database.db')
```

```python
Df1.to_sql('dataset1', conn, if_exists='replace', index=False)

```

```python
Df2.to_sql('dataset2', conn, if_exists='replace', index=False)
```

Table 1: Sno, MovieName, Director Name, Duration, genre, ratings
    
1.Display all the details of movies created by directors Christopher and Matt Reeves.

```python
table1=conn.execute("SELECT * FROM dataset1 WHERE Director1 IN ('Christopher Nolan ', 'Matt Reeves ');")
for i in table1:
    print(i)
```

2) Display all the details of movies with a duration of 140 minutes to 190 minutes.


```python
table2=conn.execute("SELECT * FROM dataset1 WHERE Time >= 140 AND time <= 190;")
for i in table2:
    print(i)
```

3) Display all details of movies with ratings above 7 in ascending order.

```python
table3=conn.execute("SELECT * FROM dataset1 WHERE Rate > 7 ORDER BY Rate ASC;")
for i in table3:
    print(i)
```

4) Display all movie names in descending order.

```python
table4=conn.execute("SELECT Title FROM dataset1 ORDER BY Title DESC; ")
for i in table4:
    print(i)
```

5) Display movie name starts with ‘P’ and their rating is greater than 7.

```python
table5=conn.execute("SELECT Title FROM dataset1 WHERE Title LIKE 'P%' AND Rate > 7;")
for i in table5:
    print(i)
```

# subquery

1) Display all details from both tables where movie names are the same.

```python
subq1= conn.execute("SELECT * FROM dataset1 JOIN dataset2 ON dataset1.Title = dataset2.Name;")

for i in subq1:
    print(i)
    
```

2. Display all movie names, Director, ratings, and gross collection where the genre is action.

```python
subq2=conn.execute("SELECT dataset1.Title, dataset1.Director1, dataset1.Rate, dataset2.Collection, dataset2.Genre1 FROM dataset1, dataset2 WHERE (dataset2.Genre1='Action' ) AND dataset1.Title=dataset2.Name;")
for i in subq2:
    print(i)
```

3. Display all the details from both tables with the highest gross collection.

```python
subq3=conn.execute("SELECT*FROM dataset1, dataset2 WHERE dataset2.Collection=(SELECT MAX(Collection) FROM dataset2) AND dataset1.Title=dataset2.Name;")
for i in subq3:
    print(i)
```

4. Display all the details from both the tables with the highest ratings.

```python
subq4=conn.execute("SELECT*FROM dataset1, dataset2 WHERE dataset1.Rate=(SELECT MAX(Rate) FROM dataset1) AND dataset1.Title=dataset2.Name;")
for i in subq4:
    print(i)
```

5.Display all details from both tables with the lowest gross collection and lowest ratings





```python
subq5=conn.execute("SELECT*FROM dataset1,dataset2 WHERE(dataset2.Collection=(SELECT min(Collection) FROM dataset2) OR dataset1.Rate=(SELECT min(Rate) FROM dataset1)) AND dataset1.Title=dataset2.Name;")
for i in subq5:
    print(i)
```
